### Data Wrangling 4
Data Wrangling strategy for answering your research questions

4.  Determine average salary by job level.

```python
# Transform with Group By
# Add a new column 'average salary per rank' into the dataframe
# Answer: Average salary per rank

df['average salary per rank'] = df.groupby('rank')['salary'].transform('mean')
```

```python
df
```

### Data Wrangling 5
Data Wrangling strategy for answering your research questions

5.  Determine if each worker is earning above or below average salary for their level.

```python
# Use another transform formula
# Add a new column field above_below average
# Answer: Any with a positive number means above average, while a negative number means below average.

df['above_below_average'] = df.groupby('rank')['salary'].transform(lambda x: x - x.mean())
```

```python
df
```
